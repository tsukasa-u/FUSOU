```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import fusou_datasets

def load_table(name: str, master: bool = False):
    try:
        return fusou_datasets.load_master(name) if master else fusou_datasets.load(name, offline=True)
    except Exception:
        return fusou_datasets.load_master(name) if master else fusou_datasets.load(name, offline=False)

fusou_datasets.configure(cache_dir="~/.fusou_datasets/cache")

try:
    df = load_table("hougeki")
except Exception:
    df = pd.DataFrame()
    print("hougeki table not available")

needed = {"hit", "damage"}
if df.empty or not needed.issubset(df.columns):
    print("hougeki table needs hit and damage columns")
else:
    bins = [0, 10, 30, 60, 100, 200, 400, 800]
    df["damage_bin"] = pd.cut(df["damage"], bins=bins, labels=["0-10", "10-30", "30-60", "60-100", "100-200", "200-400", "400+"])
    pivot = df.pivot_table(index="hit", columns="damage_bin", values="damage", aggfunc="count", fill_value=0)
    plt.figure(figsize=(10, 6))
    plt.imshow(pivot, aspect="auto", cmap="YlGnBu")
    plt.xticks(ticks=range(len(pivot.columns)), labels=pivot.columns, rotation=45)
    plt.yticks(ticks=range(len(pivot.index)), labels=pivot.index)
    plt.colorbar(label="count")
    plt.title("Hougeki Hit vs Damage Bucket")
    plt.tight_layout()
    plt.show()
```
